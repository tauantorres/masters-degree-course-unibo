```python
from minizinc import Instance, Model, Solver

# Load n-Queens model from file
nqueens = Model("./nqueens.mzn")
# Find the MiniZinc solver configuration for coin-bc
coin_bc = Solver.lookup("gecode")
# Create an Instance of the n-Queens model for coin-bc
instance = Instance(coin_bc, nqueens)
# Assign 4 to n
instance["n"] = 8
import asyncio

async def solve_instance():
	result = await instance.solve_async()
	# Output the array q
	print(result["q"])

# Run the async function
await solve_instance()

```
